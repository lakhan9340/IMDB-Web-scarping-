```python
import selenium 
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
```

```python

driver = webdriver.Chrome()
wait= WebDriverWait(driver,10,poll_frequency=1)
driver.get('https://www.imdb.com/chart/top/')
time.sleep(5)
driver.maximize_window()
```

```python
all_elements = driver.find_elements(By.CSS_SELECTOR ,"div.ipc-metadata-list-summary-item__c")
data=[]
for all in all_elements:
    data.append(all.text)
print(data)
```

```python
import pandas as pd
df=pd.DataFrame(data)
df
```

```python
df[['Movie Name', 'Year','Duration','Certificate','Ratin','View','Rate']] = df[0].str.split('\n',expand=True)
df
```

```python
df.drop(columns=[0],inplace=True)
```

```python
df.drop(columns=['Rate'],inplace=True)
```

```python
df=df.rename(columns={'Ratin':'Rating','Certificate':'Grade','View':'Rating By Peoples'})
```

```python
df
```

```python


# Remove a specific suffix from the right
df['Rating By Peoples'] = df['Rating By Peoples'].str.rstrip(')')
```

```python
df['Movie Name'] = df['Movie Name'].str[3:]
```

```python
df['Rating By Peoples'] = df['Rating By Peoples'].str.strip()
```

```python
df['Rating By Peoples'] = df['Rating By Peoples'].str.lstrip('(')
```
